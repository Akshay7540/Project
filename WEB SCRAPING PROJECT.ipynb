```python
productname = [] 
price = []
rating = []
features = []
pagenum = []

for i in range(1,25): # go throught the 23 pages
    start_time = time.time()
    URL = 'https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'.format(i)
    page = requests.get(URL)
    pagecontent = page.text
    soup = BeautifulSoup(pagecontent)

    for x in soup.find_all('div', attrs={'class':'_1-2Iqu row'}): # get 25 product properties in each page
        pname = x.find('div', attrs = {'class':'_3wU53n'})
        cost = x.find('div', attrs = {'class':'_1vC4OE _2rQ-NK'})
        rat = x.find('div', attrs = {'class':'hGSR34'})
        specs = x.find('ul', attrs = {'class':'vFw0gD'})
        
        if pname is None:
            productname.append(np.NaN)
        else:
            productname.append(pname.text)
            
        if cost is None:
            price.append(np.NaN)
        else:
            price.append(cost.text)
            
        if rat is None:
            rating.append(np.NaN)
        else:
            rating.append(rat.text)
            
        if specs is None:
            features.append(np.NaN)
        else:
            features.append(specs.text)
            
        pagenum.append(i)
    print('Page {} completed in {} seconds'.format(i, time.time()-start_time))
```

```python
mobile_df = pd.DataFrame({'ProductName':productname, 'Price':price, 'Rating': rating, 'Features':features, 'PageNo': pagenum})
```

# Extracting Features - Specification, RAM, Screensize, Brand and cleaning other data


```python
mobile_df['Features'][0]

mobile_df['Features'][1]

mobile_df['Features'][2]
```

```python
regex = r'\s\w+[0-9]\s\w+\s'

re.compile(regex).findall(mobile_df['Features'][0])

mobile_df['specification'] = mobile_df['Features'].apply(lambda x:re.compile(regex).findall(x))
```

```python
mobile_df
```

```python
mobile_df.specification.value_counts()
```

```python
regex = r'[0-9.\s]+(?:cm|inch)'

mobile_df['ScreenSize'] = mobile_df['Features'].apply(lambda x:re.compile(regex).findall(x))
```

```python
mobile_df
```

```python
regex = r'[0-9][\sA-Za-z0-9]+RAM'

mobile_df['RAM'] = mobile_df['Features'].apply(lambda x:re.compile(regex).findall(x))
```

```python
mobile_df
```

```python
regex = r'^\w+'

mobile_df['Brand'] = mobile_df['ProductName'].apply(lambda x:re.compile(regex).findall(x))
```

# Data Visualizations


### Univariate analysis:- provides summary statistics for each field in the raw data set (or) summary only on one variable.

### Bivariate analysis:- is performed to find the relationship between each variable in the dataset and the target variable of interest (or) using 2 variables and finding realtionship between them.Ex:-Box plot,Voilin plot.

### Multivariate analysis:- is performed to understand interactions between different fields in the dataset (or) finding interactions between variables more than 2. Ex:- scatter plot.
